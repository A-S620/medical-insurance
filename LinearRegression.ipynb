```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

df = pd.read_csv('insurance.csv')

model = smf.ols('charges ~ bmi', data=df)
model = model.fit()

plt.figure(figsize=(12, 6))
plt.plot(df['bmi'], df['charges'], 'o')           # scatter plot showing actual data
plt.plot(df['bmi'], model, 'r', linewidth=2)   # regression line
plt.xlabel('BMI')
plt.ylabel('Insurance Charges')
plt.title('BMI vs Charges')

plt.show()
```
